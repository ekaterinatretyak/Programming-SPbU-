```python
!pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl

import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('administration_data.csv', sep=';', engine='python', error_bad_lines=False, encoding="utf-8")
df.head()
```

```python
m1 = df[df['gender'] == 'Male'][df['age'] < 14]
f1 = df[df['gender'] == 'Female'][df['age'] < 14]

m2 = df[df['gender'] == 'Male'][(df['age'] >= 14) & (df['age'] < 20)]
f2 = df[df['gender'] == 'Female'][(df['age'] >= 14) & (df['age'] < 20)]

m3 = df[df['gender'] == 'Male'][(df['age'] >= 20) & (df['age'] < 26)]
f3 = df[df['gender'] == 'Female'][(df['age'] >= 20) & (df['age'] < 26)]

m4 = df[df['gender'] == 'Male'][(df['age'] >= 26) & (df['age'] < 32)]
f4 = df[df['gender'] == 'Female'][(df['age'] >= 26) & (df['age'] < 32)]

m5 = df[df['gender'] == 'Male'][(df['age'] >= 32) & (df['age'] <= 36)]
f5 = df[df['gender'] == 'Female'][(df['age'] >= 32) & (df['age'] < 36)]
```

```python
fig = plt.figure(figsize=(36,24), frameon=True)
plt.grid(axis='y', alpha=0.3)
spec = mpl.gridspec.GridSpec(ncols=3, nrows=2)
plt.axis("off")

ax1 = fig.add_subplot(spec[0,0])
ax1.plot(m1['age'], m1.sort_values(by='comprehension')['comprehension'], '-b', label='boys - comprehension')
ax1.plot(m1['age'], m1.sort_values(by='production')['production'], ':b', label='boys - production')
ax1.plot(f1['age'], f1.sort_values(by='comprehension')['comprehension'], '-r', label='girls - comprehension')
ax1.plot(f1['age'], f1.sort_values(by='production')['production'], ':r', label='girls - production')
ax1.set_title('6-13 months', fontsize=18)
ax1.set_xlabel('Age', fontsize=18)
ax1.set_ylabel('Number of words', fontsize=18)

plt.xticks(fontsize=20)
y_ticks = np.arange(0, 500, 100)
plt.yticks(y_ticks, fontsize=20)

plt.legend(fontsize=20, loc = 'upper left')

ax2 = fig.add_subplot(spec[0,1])
ax2.plot(m2['age'], m2.sort_values(by='comprehension')['comprehension'], '-b', label='boys - comprehension')
ax2.plot(m2['age'], m2.sort_values(by='production')['production'], ':b', label='boys - production')
ax2.plot(f2['age'], f2.sort_values(by='comprehension')['comprehension'], '-r', label='girls - comprehension')
ax2.plot(f2['age'], f2.sort_values(by='production')['production'], ':r', label='girls - production')
ax2.set_title('14-19 months', fontsize=18)
ax2.set_xlabel('Age', fontsize=18)
ax2.set_ylabel('Number of words', fontsize=18)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

ax3 = fig.add_subplot(spec[0,2])
ax3.plot(m3['age'], m3.sort_values(by='comprehension')['comprehension'], '-b', label='boys - comprehension')
ax3.plot(m3['age'], m3.sort_values(by='production')['production'], ':b', label='boys - production')
ax3.plot(f3['age'], f3.sort_values(by='comprehension')['comprehension'], '-r', label='girls - comprehension')
ax3.plot(f3['age'], f3.sort_values(by='production')['production'], ':r', label='girls - production')
ax3.set_title('20-25 months', fontsize=18)
ax3.set_xlabel('Age', fontsize=18)
ax3.set_ylabel('Number of words', fontsize=18)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

ax4 = fig.add_subplot(spec[1,0])
ax4.plot(m4['age'], m4.sort_values(by='comprehension')['comprehension'], '-b', label='boys - comprehension')
ax4.plot(m4['age'], m4.sort_values(by='production')['production'], ':b', label='boys - production')
ax4.plot(f4['age'], f4.sort_values(by='comprehension')['comprehension'], '-r', label='girls - comprehension')
ax4.plot(f4['age'], f4.sort_values(by='production')['production'], ':r', label='girls - production')
ax4.set_title('26-31 months', fontsize=18)
ax4.set_xlabel('Age', fontsize=18)
ax4.set_ylabel('Number of words', fontsize=18)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

ax5 = fig.add_subplot(spec[1,2])
ax5.plot(m5['age'], m5.sort_values(by='comprehension')['comprehension'], '-b', label='boys - comprehension')
ax5.plot(m5['age'], m5.sort_values(by='production')['production'], ':b', label='boys - production')
ax5.plot(f5['age'], f5.sort_values(by='comprehension')['comprehension'], '-r', label='girls - comprehension')
ax5.plot(f5['age'], f5.sort_values(by='production')['production'], ':r', label='girls - production')
ax5.set_title('32-36 months', fontsize=18)
ax5.set_xlabel('Age', fontsize=18)
ax5.set_ylabel('Number of words', fontsize=18)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.show()
```
